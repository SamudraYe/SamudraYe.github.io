```python
# 导入 opencv 库
import cv2

# 获取OpenCV版本号
print(cv2.__version__)
```

#### <span id="1">[1 掌握 cv2.imread()读取图像](#t1)</span>

```python
# imread(filename[, flags]),
# param1：文件路径，若路径错误，opencv并不会报错，只是在print(img)时输出None
# param2：默认是以RGB模式处理图片
#        -1 = IMREAD_UNCHANGED
#         0 = IMREAD_GRAYSCALE
#         1 = IMREAD_COLOR
#         2 = IMREAD_ANYDEPTH
#         3 = IMREAD_COLOR
#         4 = IMREAD_ANYCOLOR
#        16 = IMREAD_REDUCED_GRAYSCALE_2
#        32 = IMREAD_REDUCED_GRAYSCALE_4
#        64 = IMREAD_REDUCED_GRAYSCALE_2
#        17 = IMREAD_REDUCED_COLOR_2
#        33 = IMREAD_REDUCED_COLOR_4
#        65 = IMREAD_REDUCED_COLOR_8
#       128 = IMREAD_IGNORE_ORIENTATION

img = cv2.imread("./pics/2017-02-13_12.jpg")
```

#### <span id="2">[2 掌握 imshow()显示图像](#t2)</span>

```python
# imshow(winname, mat)
# param1：窗口名称，注意窗口会自动调整为图像大小
# param2：imread读取的图像
cv2.imshow("2017-02-13_12",img)

# 若要创建多个窗口，必须命名不同的 winname
cv2.imshow("2017-02-13_12-1",img)
cv2.imshow("2017-02-13_12-2",img)
```

#### <span id="2.1">[2.1 cv2.waitKey](#t2.1)</span>

```python
# waitKey([, delay])
# 说明：键盘绑定函数，其时间尺度是毫秒级
# 功用：函数等待特定的几毫米，看是否有键盘输入；
# 返回值：在设定的时间内，按下任意键，函数会返回按键的ASCII值；
#       如果没有键盘输入，返回值为 -1；
#       若设定参数为0，表示无限期的等待键盘输入。

cv2.imshow("2017-02-13_12",img)

while True:
    key = cv2.waitKey(0)
    
    if 27 == key: # wait for "ESC" key to exit
        cv2.destroyAllWindows()
        break
    else: # 打印输入的键值
        msg = "key {}({}) is pressed.".format(chr(key), key)
        print(msg)
```

#### <span id="2.2">[2.2 cv2.destroyAllWindows](#t2.2)</span>

```python
cv2.imshow("2017-02-13_12",img)
cv2.imshow("2017-02-13_12-1",img)
cv2.imshow("2017-02-13_12-2",img)

while True:
    key = cv2.waitKey(0)

    if 27 == key: # wait for "ESC" key to exit
        cv2.destroyAllWindows() # 删除所有窗口
        break
    elif key == ord('d'):
        cv2.destroyWindow("2017-02-13_12-1") # 删除指定窗口
    else:
        cv2.destroyWindow("2017-02-13_12-2") # 删除指定窗口
```

#### <span id="2.3">[2.3 cv2.namedWindow](#t2.3)</span>

```python
# namedWindow(winname[, flags])
# param1: windows name
# param2: flags
#         0 = WINDOW_NORMAL
#         0 = WINDOW_GUI_EXPANDED
#         0 = WINDOW_KEEPRATIO
#         1 = WINDOW_AUTOSIZE
#         1 = WINDOW_FULLSCREEN
#        16 = WINDOW_GUI_NORMAL
#       256 = WINDOW_FREERATIO
#      4096 = WINDOW_OPENGL

cv2.namedWindow("girl", 0)
cv2.imshow("girl",img)

if 27 == cv2.waitKey(0):# wait for "ESC" key to exit
    cv2.destroyAllWindows()
```
